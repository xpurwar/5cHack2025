```python
# Getting all menus for Hoch Dining Hall
# first 7 days, all meals

driver = webdriver.Chrome()
driver.get("https://hmc.sodexomyway.com/en-us/locations/hoch-shanahan-dining-commons")
wait = WebDriverWait(driver, 10)


hoch_menus = []

calendar = wait.until(
    EC.presence_of_element_located((By.CLASS_NAME, "monthContainer"))
)
month_text = calendar.find_element(By.CLASS_NAME, "monthLabel").text
month_num = month_to_num[month_text]
# print(f"month: {month}")

date_buttons = calendar.find_elements(By.CLASS_NAME, "dateDayItem")

limit = 7
i = 0
# for each day
for date_button in date_buttons:
    i += 1
    if i > limit: break
    day_text = date_button.find_element(By.CLASS_NAME, "dayLabel").text
    date_text = date_button.find_element(By.CLASS_NAME, "dateLabel").text
    date_button.click()
    time.sleep(0.5)

    # get meal access
    meal_section = wait.until(
        EC.presence_of_element_located((By.CLASS_NAME, "Menustyles__MenuComponent-sc-eo2633-1"))
    )
    meals = meal_section.find_elements(By.CLASS_NAME, "Menustyles__MenuMealSection-sc-eo2633-2")

    # for each meal
    for meal in meals:
        menu = {}
        menu["dining_hall"] = "Hoch"
        menu["date"] = f"2025-{month_num}-{date_text}"
        menu["day"] = day_text

        # get meal text (breakfast, brunch, lunch, dinner)
        meal_text = meal.find_element(By.CLASS_NAME, "header").text.lower()
        menu["meal"] = meal_text

        # open meal
        meal_button = meal.find_element(By.CSS_SELECTOR, 'div[role="button"]')
        meal_button.click()
        time.sleep(0.5)

        # get stations
        menu["stations"] = {}
        stations = meal.find_elements(By.CSS_SELECTOR, ".MenuCourseSectionstyles__MenuComponent-sc-12v9ie3-0.dVaXYi")

        for station in stations:
            station_text = station.find_element(By.CSS_SELECTOR, "h3").text.lower()
            if station_text:
                menu["stations"][station_text] = []

                foods = station.find_elements(By.CSS_SELECTOR, ".menu-item-container")
                for food in foods:
                    food_dict = {}

                    name_text = food.find_element(By.CSS_SELECTOR, '.name').text
                    cal_text = food.find_element(By.CSS_SELECTOR, '.calories').text

                    # food restrictions 
                    allergens = []
                    diets = []
                    icons = food.find_elements(By.CSS_SELECTOR, "img")
                    for icon in icons:
                        icon_text = icon.get_attribute("aria-label")
                        if icon_text in all_allergens: allergens.append(icon_text)
                        if icon_text in all_diets: diets.append(icon_text)

                    food_dict["name"] = name_text
                    food_dict["cal"] = cal_text
                    food_dict["allergens"] = allergens
                    food_dict["diets"] = diets
                    menu["stations"][station_text].append(food_dict)
            
        # add menu to hoch_menus
        hoch_menus.append(menu)

        # close meal
        meal_button.click()
        time.sleep(0.5)

with open('hoch_menus.json', 'w', encoding='utf-8') as f:
    json.dump(hoch_menus, f, ensure_ascii=False, indent=4)


driver.close()
driver.quit()
```

```python
# Getting all food items for Hoch Dining Hall for Ananya
# first 7 days, all meals

driver = webdriver.Chrome()
driver.get("https://hmc.sodexomyway.com/en-us/locations/hoch-shanahan-dining-commons")
wait = WebDriverWait(driver, 10)


hoch_food_items = []

calendar = wait.until(
    EC.presence_of_element_located((By.CLASS_NAME, "monthContainer"))
)
month_text = calendar.find_element(By.CLASS_NAME, "monthLabel").text
month_num = month_to_num[month_text]
# print(f"month: {month}")

date_buttons = calendar.find_elements(By.CLASS_NAME, "dateDayItem")

limit = 7
i = 0
# for each day
for date_button in date_buttons:
    i += 1
    if i > limit: break
    day_text = date_button.find_element(By.CLASS_NAME, "dayLabel").text
    date_text = date_button.find_element(By.CLASS_NAME, "dateLabel").text
    date_button.click()
    time.sleep(0.8)

    # get meal access
    meal_section = wait.until(
        EC.presence_of_element_located((By.CLASS_NAME, "Menustyles__MenuComponent-sc-eo2633-1"))
    )
    meals = meal_section.find_elements(By.CLASS_NAME, "Menustyles__MenuMealSection-sc-eo2633-2")

    # for each meal
    for meal in meals:
        

        # get meal text (breakfast, brunch, lunch, dinner)
        meal_text = meal.find_element(By.CLASS_NAME, "header").text.lower()
        menu["meal"] = meal_text

        # open meal
        meal_button = meal.find_element(By.CSS_SELECTOR, 'div[role="button"]')
        meal_button.click()
        time.sleep(0.8)

        # get stations
        menu["stations"] = {}
        stations = meal.find_elements(By.CSS_SELECTOR, ".MenuCourseSectionstyles__MenuComponent-sc-12v9ie3-0.dVaXYi")

        for station in stations:
            station_text = station.find_element(By.CSS_SELECTOR, "h3").text.lower()
            if station_text:
                foods = station.find_elements(By.CSS_SELECTOR, ".menu-item-container")
                for food in foods:
                    # food name and calories
                    name_text = food.find_element(By.CSS_SELECTOR, '.name').text
                    cal_text = food.find_element(By.CSS_SELECTOR, '.calories').text
                    # food restrictions 
                    allergens = []
                    diets = []
                    icons = food.find_elements(By.CSS_SELECTOR, "img")
                    for icon in icons:
                        icon_text = icon.get_attribute("aria-label")
                        if icon_text in all_allergens: allergens.append(icon_text)
                        if icon_text in all_diets: diets.append(icon_text)

                    food_item = {}
                    food_item["dining_hall"] = "Hoch"
                    food_item["date"] = f"2025-{month_num}-{date_text}"
                    food_item["day"] = day_text
                    food_item["meal"] = meal_text
                    food_item["station"] = station_text
                    food_item["name"] = name_text
                    food_item["cal"] = cal_text
                    if allergens: food_item["allergens"] = allergens
                    else: food_item["allergens"] = None
                    if diets: food_item["diets"] = diets
                    else: food_item["diets"] = None

                    hoch_food_items.append(food_item)

        # close meal
        meal_button.click()
        time.sleep(0.8)

with open('hoch_food_items.json', 'w', encoding='utf-8') as f:
    json.dump(hoch_food_items, f, ensure_ascii=False, indent=4)


driver.close()
driver.quit()
```
